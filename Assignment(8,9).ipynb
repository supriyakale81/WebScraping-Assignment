### 8) Write a python program to scrape the details of most downloaded articles from AI in last 90days. https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
#### Scrape below mentioned details and make data frame
##### i) Paper Title
##### ii) Authors
##### iii) Published Date
##### iv) Paper URL


```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
headlines=[]
headline=soup.find_all('h1')
for i in headline:
    headlines.append(i.text)
```

```python
headlines
```

```python
titles=[]
title=soup.find_all('h2')
for i in title:
    titles.append(i.text)
```

```python
print(titles)
```

```python
authors=[]
author=soup.find_all('span',class_="sc-1w3fpd7-0 dnCnAO")
for i in author:
    authors.append(i.text)
```

```python
print(authors)
```

```python
dates=[]
date=soup.find_all('span',class_="sc-1thf9ly-2 dvggWt")
for i in date:
    dates.append(i.text)
```

```python
print(dates)
```

```python
urls=[]
url=soup.find_all("a",class_="sc-5smygv-0 fIXTHm")
for i in url:
    urls.append(i['href'])
```

```python
print(urls)
```

```python
print(len(titles),len(authors),len(dates),len(urls))
```

```python
import pandas as pd
data=pd.DataFrame({"Paper Title":titles,"Authors":authors,"Published Date":dates,"Paper URL":urls})
data
```

### 9) Write a python program to scrape mentioned details from dineout.co.in and make data frame
#### i) Restaurant name
#### ii) Cuisine
#### iii) Location
#### iv) Ratings
#### v) Image URL

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
names=[]
name=soup.find_all('a',class_="restnt-name ellipsis")
for i in name:
    names.append(i.text)
```

```python
print(names)
```

```python
cus=[]
cu=soup.find_all('span',class_="double-line-ellipsis")
for i in cu:
    cus.append(i.text.split('|')[-1])
    
```

```python
print(cus)
```

```python
locations=[]
location=soup.find_all('div',class_="restnt-loc ellipsis")
for i in location:
    locations.append(i.text.split(",")[-1])
```

```python
print(locations)
```

```python
ratings=[]
rating=soup.find_all('div',class_="restnt-rating rating-4")
for i in rating:
    ratings.append(i.text)
```

```python
print(ratings)
```

```python
urls=[]
url=soup.find_all("img",class_="no-img")
for i in url:
    urls.append(i.get('data-src'))
```

```python
print(urls)
```

```python
print(len(names),len(locations),len(cus),len(urls))
```

```python
import pandas as pd
data=pd.DataFrame({"Restaurant name":names,"Cuisine":cus,"Location":locations,"Ratings":ratings,"Image URL":urls})
data
```
