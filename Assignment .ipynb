### 1) Write a python program to display all the header tags from wikipedia.org and make data frame.


```python
# Python program to print all heading tags
import requests
from bs4 import BeautifulSoup

# scraping a wikipedia article
url_link = "https://en.wikipedia.org/"
request = requests.get(url_link)

Soup = BeautifulSoup(request.text, 'lxml')

# creating a list of all common heading tags
heading_tags = ["h1", "h2", "h3","h4"]
for tags in Soup.find_all(heading_tags):
    print(tags.name + ' -> ' + tags.text.strip())
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
url = "https://en.wikipedia.org/"
fileObj = urlopen(url)
html = fileObj.read()
```

```python
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
headerlist = soup.find_all(re.compile('^h[1-6]'))
print("Headers : ",headerlist)
```

```python
headers_tag=[]
headers_text=[]
for header in headerlist:
    headers_tag.append(header.name)
    headers_text.append(header.text)
print(headers_tag)
print(headers_text)
```

```python
import pandas as pd
df=pd.DataFrame({"Tag Name":headers_tag,"Headers Text":headers_text})
df
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# connect to the driver
driver = webdriver.Chrome(r"C:\Users\pc\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
# scraping the naukri page automated chrome browser
driver.get("https://imdb.com")
```

```python
menu=driver.find_element(By.ID,"iconContext-menu")
menu.click()
```

```python
movies=driver.find_element(By.XPATH,'//label[@class="navlinkcat__item"]')
movies.click()
```

```python
top=driver.find_element(By.XPATH,'//span[@class="ipc-list-item__text"]')
top.click()
```

```python
title=driver.find_element(By.CLASS_NAME,'header')
title.text
```

```python
year=driver.find_element(By.CLASS_NAME,"secondaryInfo")
year.text
```

```python
rating=driver.find_element(By.TAG_NAME,"strong")
rating.text
```

```python
names=[]
name=driver.find_elements(By.XPATH,'//td[@class="titleColumn"]')
for i in name[0:50]:
    names.append(i.text[3:-6])
print(names)
# ab=names[:-6]
# print(ab)
```

```python
years=[]
year=driver.find_elements(By.CLASS_NAME,"secondaryInfo")
for i in year[0:50]:
    years.append(i.text[1:-1])
print(years)
    
```

```python
ratings=[]
rating=driver.find_elements(By.TAG_NAME,"strong")
for i in rating[0:50]:
    ratings.append(i.text)
print(ratings)
    
```

```python
print(len(names),len(years),len(ratings))
```

```python
import pandas as pd
df=pd.DataFrame({"Movies Name":names,"Released Year":years,"Movie Ratings":ratings})
df
```
