```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("csv1(raw).csv")
df
```

```python
df['Year'].unique()
```

```python
df['directors_list'] = df['Director'].str.split(',')
```

```python
df_direc = pd.DataFrame(df['directors_list'].tolist()).fillna('').add_prefix('Director_')
# df.add_prefix('Sold_')
df_direc=df_direc.rename(columns = {'Director_0':'Director1','Director_1':'Director2','Director_2':'Director3','Director_3':'Director4','Director_4':'Director5','Director_5':'Director6'})
df_direc
```

```python
df = pd.concat([df, df_direc], axis=1)

df
```

```python
df.drop(['directors_list','Director'],axis=1,inplace=True)
```

```python
df.to_csv("CSV1_normalized.csv",index=False)
```
