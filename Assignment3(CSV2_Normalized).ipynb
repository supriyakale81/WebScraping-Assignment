```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("csv2(raw).csv")
df.head()
```

```python
df.shape
```

```python
df['stars_list'] = df['Stars'].str.split(',')
```

```python
df_stars = pd.DataFrame(df['stars_list'].tolist()).fillna('').add_prefix('Stars_')

df_stars=df_stars.rename(columns={"Stars_0":'Star1',"Stars_1":"Star2","Stars_2":"Star3","Stars_3":"Star4"})
```

```python
df_stars
```

```python
df['genre_list'] = df['Genre'].str.split(',')
```

```python
df_genre = pd.DataFrame(df['genre_list'].tolist()).fillna('').add_prefix('Genre_')

df_genre=df_genre.rename(columns={"Genre_0":"Genre1","Genre_1":"Genre2","Genre_2":"Genre3"})
df_genre
```

```python
df = pd.concat([df, df_stars,df_genre], axis=1)

df
```

```python
df.drop(['stars_list','genre_list','Stars','Genre'],axis=1,inplace=True)
```

```python
df.to_csv("CSV2_normalized.csv",index=False)
```
