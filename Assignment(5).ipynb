```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# connect to the driver
driver = webdriver.Chrome(r"C:\Users\pc\Downloads\chromedriver_win32\chromedriver.exe")

# scraping the naukri page automated chrome browser
driver.get("https://www.icc-cricket.com/rankings/mens/player-rankings/test")
```

```python
search=driver.find_element(By.XPATH,'/html/body/div[2]/header/section[2]/div[2]/div/nav/a[3]')
search.click()
```

```python
Players=[]
ratings=[]
teams=[]
```

```python
rate=driver.find_elements(By.XPATH,'//td[@class="table-body__cell u-text-right rating"]')
for i in rate[0:10]:
    ratings.append(i.text)
print(ratings)
```
