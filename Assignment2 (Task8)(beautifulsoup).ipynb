### Write a python program to scrape data for Top 1000 Quotes of All Time.
#### The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuotes
3. Than scrap a) Quote b) Author c) Type Of Quote

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("http://www.azquotes.com/")
page
```

```python
soup=BeautifulSoup(page.content,"html.parser")
print(soup)
```

```python
TopQ=[]
top=soup.find_all("a",class_="title")
for i in top:
    TopQ.append(i.text)
```

```python
print(TopQ)
```

```python
author=[]
A=soup.find_all('div',class_="author")
for i in A:
    author.append(i.get_attribute_list)
```

```python
print(author)
```

```python
Type=[]
t=soup.find_all('div',class_="tags")
for i in t:
    Type.append(i.text)
```
